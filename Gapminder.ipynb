```python
merged_countries = pd.concat([df_gini['country'], df_pcap['country']]).drop_duplicates().reset_index(drop=True)

# Create a new DataFrame with the merged 'country' column
merged_df = pd.DataFrame({'country': merged_countries})

# Print the new DataFrame
print(merged_df)
```

```python
population_1900 = df_pop[['country', '1900']]
income_1900 = df_pcap[['country', '1900']]
gini_1900 = df_gini[['country', '1900']]

# Merge dataframes based on country names
merged_data = pd.merge(population_1900, income_1900, on='country')
merged_data = pd.merge(merged_data, gini_1900, on='country')
merged_data = merged_data.rename({'1900_x': 'population', '1900_y': 'income_pcap', '1900': 'gini'}, axis=1)

merged_data.dtypes
```

```python
def convert_population(population_str):
    if 'M' in population_str:
        # If population is in millions, remove 'M' and convert to integer
        return int(float(population_str.replace('M', '')) * 1_000_000)
    elif 'k' in population_str:
        # If population is in thousands, remove 'K' and convert to integer
        return int(float(population_str.replace('k', '')) * 1_000)
    else:
        # If population is already an integer, return it
        return int(population_str)

# Example usage:
population_int = convert_population("3.5M")  # This will return 3500000
```

```python
merged_data['population'] = merged_data['population'].apply(convert_population)
```

```python
merged_data['population']
```

```python
# Plot scatter plot
plt.figure(figsize=(40, 20))

# Scatter plot with population as bubble size
plt.scatter(merged_data['gini'], merged_data['income_pcap'], s=merged_data['population']*0.0001, c=range(len(merged_data)), cmap='viridis', alpha=0.7)

# Adding labels and title
plt.xlabel('Gini Coefficient (1900)')
plt.ylabel('Income per Capita (1900)')
plt.title('Scatter Plot: Gini Coefficient vs Income per Capita (1900)')

for i, txt in enumerate(merged_data['country']):
    plt.annotate(txt, (merged_data['gini'][i], merged_data['income_pcap'][i]), textcoords="offset points", xytext=(0,10), ha='center')


# Show plot
plt.grid(True)
plt.show()
```
